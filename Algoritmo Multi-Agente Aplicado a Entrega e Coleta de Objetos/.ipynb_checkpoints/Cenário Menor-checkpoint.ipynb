```python
a_max = np.argmax(Rewards)
print(f'melhor recompensa real: {Rewards[a_max]}')
print(f'melhor recompensa encontrada: {bests_rewards[-1]}\n')
print('sendo que essas sequ√™ncias s√£o')

dim = len(best_steps)
c = np.zeros([dim, 4])

for i in range(dim):
    for j in range(2):
        c[i,j] = caminhos[a_max][i][j]
        c[i,j+2] = best_steps[i][j]

pd.DataFrame(c, columns=['A0-R', 'A1-R', 'A0-E', 'A1-E'])
```

```python
rotas = [[0 for i in range(n_agents)] for j in range(len(best_steps)-1)]

for i in range(1,len(best_steps)):
    for j in range(n_agents):
        if best_steps[i][j]>=n_agents and best_steps[i][j]<(n_agents+n_objects):
            rotas[i-1][j] = "O"+str(best_steps[i][j]-n_agents)
        else:
            if best_steps[i][j] != 0:
                rotas[i-1][j] = "R"+str(best_steps[i][j]-n_agents-n_objects)

data = pd.DataFrame(rotas)
data.head(len(rotas))
```
